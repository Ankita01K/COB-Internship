# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# List of columns to label encode
columns_to_encode = ['type', 'title', 'director', 'country', 'release_year', 'rating', 'duration', 'listed_in']
df3= df2[columns_to_encode].apply(lambda col: le.fit_transform(col))
```

```python
df3
```

# Outlier 

```python
plt.figure(figsize=(11,8))  
for i, feature in enumerate(col_name_list, 1):  # Start index from 1
    plt.subplot(4, 4, i)  
    sns.boxplot(data=df3[feature])
    plt.title(f'Box Plot for {feature}')
    # Calculate the number of outliers
    q1 = df3[feature].quantile(0.25)
    q3 = df3[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    num_outliers = ((df3[feature] < lower_bound) | (df3[feature] > upper_bound)).sum()
    plt.text(0.1,0.1,f'Outliers: {num_outliers}', transform=plt.gca().transAxes)
    plt.title(f'Box Plot for {feature}')
plt.tight_layout()
plt.show()
```

## The 'release_year' column has 71 outliers,and 'rating' column has 13 outliers.

# Remove outliers from these columns using the Interquartile Range (IQR) method

```python
q1 = df3[['release_year', 'rating']].quantile(0.25)
q3 = df3[['release_year', 'rating']].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers for the 'release_year' and 'rating' columns
outliers = ((df3['release_year'] < lower_bound['release_year']) | (df3['release_year'] > upper_bound['release_year']) |
            (df3['rating'] < lower_bound['rating']) | (df3['rating'] > upper_bound['rating']))

# Remove rows with outliers
df4= df3[~outliers]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

X = data[['x']]
Y = data[['y']]
```

```python

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=44)

print("Shape of X_train", X_train.shape)
print("Shape of y_train", y_train.shape)
print("Shape of X_test", X_test.shape)
print("Shape of y_test", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn import metrics

lr=LinearRegression()
lr.fit(X_train,y_train)
print("Intercept for the model is ",lr.intercept_,"and the scope is ",lr.coef_)

y_pred=lr.predict(X_test)
#Actual and Predicted values 
print("Predicted values:",y_pred[0:5])
print("Actual values:",Y.values[0:5])
```

```python
MAE = metrics.mean_absolute_error(y_test, y_pred)
print("Linear Regression MAE:{}".format(MAE))

MSE = metrics.mean_squared_error(y_test, y_pred)
print('Linear Regression MSE: {}'.format(np.sqrt(MSE)))

R2 = metrics.r2_score(y_test, y_pred)
print("R2 Score:{}".format(R2))
```

###  The linear regression model's impressive performance, with low errors (MAE: 2.26, MSE: 2.82) and a high R2 score of 0.99

```python
# Convert y_test and y_pred to NumPy arrays
y_test = y_test.to_numpy().ravel()
y_pred = y_pred.ravel()
```

```python
# Create a scatter plot of actual vs. predicted values
plt.scatter(y_test, y_pred, color='blue', label='Actual vs. Predicted')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')

# Calculate the range based on the data points
y_range = np.linspace(np.min([y_test.min(), y_pred.min()]), np.max([y_test.max(), y_pred.max()]), 100)

# Plot the regression line
plt.plot(y_range, y_range, color='red', linestyle='--', lw=2, label='Regression Line')

plt.legend()
plt.title('Linear Regression: Actual vs. Predicted')
plt.show()

```
