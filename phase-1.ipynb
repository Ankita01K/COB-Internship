# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# List of columns to label encode
columns_to_encode = ['type', 'title', 'director', 'country', 'release_year', 'rating', 'duration', 'listed_in']
df4[columns_to_encode] = df4[columns_to_encode].apply(lambda col: le.fit_transform(col))
```

```python
df4
```

# Outlier 

```python
plt.figure(figsize=(11,8))  
for i, feature in enumerate(col_name_list, 1):  # Start index from 1
    plt.subplot(4, 4, i)  
    sns.boxplot(data=df4[feature])
    plt.title(f'Box Plot for {feature}')
    # Calculate the number of outliers
    q1 = df4[feature].quantile(0.25)
    q3 = df4[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    num_outliers = ((df4[feature] < lower_bound) | (df4[feature] > upper_bound)).sum()
    plt.text(0.1,0.1,f'Outliers: {num_outliers}', transform=plt.gca().transAxes)
    plt.title(f'Box Plot for {feature}')
plt.tight_layout()
plt.show()
```

## The 'release_year' column has 71 outliers,and 'rating' column has 13 outliers.

# Remove outliers from these columns using the Interquartile Range (IQR) method

```python
q1 = df4[['release_year', 'rating']].quantile(0.25)
q3 = df4[['release_year', 'rating']].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers for the 'release_year' and 'rating' columns
outliers = ((df4['release_year'] < lower_bound['release_year']) | (df4['release_year'] > upper_bound['release_year']) |
            (df4['rating'] < lower_bound['rating']) | (df4['rating'] > upper_bound['rating']))

# Remove rows with outliers
df4 = df4[~outliers]
```
